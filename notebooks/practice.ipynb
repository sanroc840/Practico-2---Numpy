# **<span style='color: blue;'>Generación de datos**

## Genera dos distribuciones aleatorias de un tamaño de 1.000 observaciones cada una:

1. Una Distribución Normal (Gaussiana) con media y desvío estándar a elección.
`np.random.normal(loc=647, scale=10, size=1000)`
2. Una Distribución Uniforme dentro de un rango que definas.
`np.random.uniform(low=1, high=1000, size=1000)`

```python
# 1
distribucion_normal = np.random.normal(647, 10, 1000)

# 2
distribucion_uniforme = np.random.uniform(low=1, high=1000, size=1000)


distribucion_normal.dtype
```

## Se viasualizan las distribuciones con un tope de 20 observaciones.

```python
distribucion_normal.sort()
distribucion_normal[:20]
```

```python
distribucion_uniforme[:20]
```

#  <span style='color: blue;'>**Análisis estadístico**

## Calcula para cada distribución los siguientes valores:


 ### Media de las distribuciones:

`mean(datos) `Obtiene la media de la distribución

```python
"Media distribución normal:", round(np.mean(distribucion_normal),3)
```

```python
"Media distribución uniforme:", round(np.mean(distribucion_uniforme),3)
```

 ### Mediana de las distribuciones:

`median(datos)` Obtiene la mediana de la distribución

```python
"Mediana distribución normal:", np.median(distribucion_normal)
```

```python
"Mediana distribución uniforme:", np.median(distribucion_uniforme)
```

### Cálculo de la varianza: `var(datos)`

```python
"varianza de dsitribución normal = ", np.var(distribucion_normal)
```

```python
"Varianza de distribución uniforme = ", np.var(distribucion_uniforme)
```

### Desvío estándar

`np.std(distribucion)` Obtiene la desviación estandar de la distribución.

```python
"Desvio estandar dsitribución normal: ",np.std(distribucion_normal)
```

```python
"Desvio Estandar distribución Uniforme: ",np.std(distribucion_uniforme)
```

### Cálculo de los valores mínimos y máximos

`min(distribucion)` Cálcula el valor mínimo de la distribución

`max(distribucion)` Cálcula el valor máximo de la distribución

Distribución Normal

```python
# Mínimo
distribucion1_minimo = min(distribucion_normal)
distribucion1_minimo
```

```python
# Máximo
distribucion1_max = max(distribucion_normal)
distribucion1_max
```

Distribución Uniforme

```python
# Mínimo
distribucion1_minimo = min(distribucion_uniforme)
distribucion1_minimo
```

```python
# Máximo
distribucion1_max = max(distribucion_uniforme)
distribucion1_max
```

# <span style='color: blue;'> **Visualización**

Se genera las gráficas los histogramas en `subplots` en este caso en filas separadas para mostrar los mismos en una mejor resolución.

**Ax1** representa el histograma de la distribución normal.

**Ax2** representa el histograma de la distribuión uniforme.

```python
# Create a 2x2 grid of subplots

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1,2,figsize=(16,6) )
ax2.grid(True)
ax1.grid(True)
ax1.hist(distribucion_normal, edgecolor='black', color='green', label="Histograma de distribución normal")
ax1.set_title('Histograma distribución normal')

ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

ax2.hist(distribucion_uniforme, edgecolor='black', label="Histograma de distribución uniforme")
ax2.legend()
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')
ax2.set_title('Histograma distribución uniforme')
f.subplots_adjust(right=0.8)
# Add a super title to the entire figure
plt.suptitle('Histogramas de distribución normal y uniforme', fontsize=16)

# Adjust spacing between subplots
ax1.legend(loc='upper left')


plt.tight_layout()
# Display the plot
```

# <span style='color: blue;'>**Conclusiones y reflexión**

En la primera distribución **'normal o gaussiana'** se da en un rango menor a la distribución uniforme. Además, sus valores media y desviación estandar son los dados en la creación de la distribución.

El rango entre su máximo y mínimo es más reducido que si lo comparo con la **'distribución uniforme'**, y con el gráfico puedo añadir que la distribución **'normal'** cuya forma es la campana de Gauss, lo que es correcto para el tipo de distribución creada, mientras que la segunda distribución presenta una forma más **uniforme** con una frecuencia de repetición de los datos más pareja en todo su rango.
